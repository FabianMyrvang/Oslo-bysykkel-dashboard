```python
# Loading packages

# For data wrangling and plotting
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.graph_objects as go
import plotly.express as px
import plotly.graph_objects as go

# For dash layout
from jupyter_dash import JupyterDash
from dash import dcc, html
from dash.dependencies import Input, Output
import dash_bootstrap_components as dbc
from dash_bootstrap_templates import load_figure_template

# For date wrangling
from datetime import datetime

# For path 
import os

# For mapping
import geopandas as gpd
from shapely.geometry import Point, Polygon
import folium
```

```python
data = []
folder_path = 'bysykkel'

for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        file_path = os.path.join(folder_path, file_name)
        
        df = pd.read_csv(file_path)
        
        data.append(df)

data = pd.concat(data, ignore_index = True)
```

```python
data = pd.DataFrame(data)
data.head()
```

```python
data.sort_values('started_at', inplace=True)
# Converting to datetime
data['started_at'] = pd.to_datetime(data['started_at'])
data['ended_at'] = pd.to_datetime(data['ended_at'])

# Removing unessasary info
data['started_at_month'] = data['started_at'].dt.strftime('%Y-%m')
data['ended_at_month'] = data['ended_at'].dt.strftime('%Y-%m')

# Adding new columns for separating the month, day, hour and minute
data['started_at_day'] = pd.to_datetime(data['started_at']).dt.day
data['ended_at_day'] = pd.to_datetime(data['ended_at']).dt.day
data['started_at_hour'] = pd.to_datetime(data['started_at']).dt.hour
data['ended_at_hour'] = pd.to_datetime(data['ended_at']).dt.hour
data['started_at_minute'] = pd.to_datetime(data['started_at']).dt.minute
data['ended_at_minute'] = pd.to_datetime(data['ended_at']).dt.minute

# Converting back to datetime for started_at and ended_at
data['started_at'] = pd.to_datetime(data['started_at']).dt.date
data['ended_at'] = pd.to_datetime(data['ended_at']).dt.date

data['started_at'] = pd.to_datetime(data['started_at'])
data['ended_at'] = pd.to_datetime(data['ended_at'])

# Oslo bysykkel states that they only are operating from 01.00 to 05.00, therefore i removed all values falling in between
data = data[~((data['started_at_hour'] >= 0) & (data['started_at_hour'] < 5))]
data.dropna(inplace = True)
data.reset_index(drop = True)
data
```

```python
def get_popular_stations_df(station_type, df = data):
    # Creating a popular station dataframe for each of the unique station_ids
    popular_stations = pd.DataFrame(df.groupby(station_type + '_station_id').size()).reset_index()
    popular_stations = popular_stations.rename(columns={0: 'station_count'})

    # Dropping duplicates on station_id
    unique_stations = df[[station_type + '_station_id',
                            station_type + '_station_name',
                            station_type + '_station_description',
                            station_type + '_station_latitude',
                            station_type + '_station_longitude']].drop_duplicates(subset=[station_type + '_station_id'])

    # Merging with total count for each station
    popular_stations = pd.merge(popular_stations, unique_stations, on=station_type + '_station_id', how='inner')
    
    return popular_stations


popular_start_stations = get_popular_stations_df("start")
popular_end_stations = get_popular_stations_df("end")
```

```python
def popular_stations(df):
    # Defining the coordinate reference system
    crs = 'epsg:4326'
    
    # Extract the column prefix dynamically
    column_prefix = '_'.join(df.columns[0].split('_')[:-1])
    
    # All geometry points
    geometry = [Point(xy) for xy in zip(df[column_prefix + '_longitude'], df[column_prefix + '_latitude'])]
    
    geodata = gpd.GeoDataFrame(df, crs=crs, geometry=geometry)
    
    # Calculate the maximum popularity count
    max_popularity = geodata['station_count'].max()
    
    # Create a folium map centered around the mean latitude and longitude
    center_lat = df[column_prefix + '_latitude'].mean()
    center_lon = df[column_prefix + '_longitude'].mean()
    
    map_oslo = folium.Map(location=[center_lat, center_lon], zoom_start=13, tiles='CartoDB Positron')
    
    # Add CircleMarkers to the map
    for idx, row in geodata.iterrows():
        radius = row['station_count'] / max_popularity * 20  # Scale the size based on popularity
        if column_prefix == 'end_station':
            color = 'red'
        elif column_prefix == 'start_station':
            color = 'blue'
        popup_text = f"Station: {row[column_prefix + '_name']}<br> Bike count: {row['station_count']}"
        folium.CircleMarker(
            location=[row[column_prefix + '_latitude'], row[column_prefix + '_longitude']],
            radius=radius,
            color= color,
            fill=True,
            fill_color= color,
            fill_opacity=0.6,
            popup=popup_text
        ).add_to(map_oslo)

    # Display the map
    return map_oslo
popular_stations(popular_start_stations)
```

```python
def get_popular_stations_month(station_type, df=data):
    # Creating a popular station dataframe for each unique station_id and month
    popular_stations = pd.DataFrame(df.groupby([station_type + '_station_id', 'started_at_month']).size()).reset_index()
    popular_stations = popular_stations.rename(columns={0: 'station_count'})

    # Dropping duplicates on station_id
    unique_stations = df[[station_type + '_station_id',
                          station_type + '_station_name',
                          station_type + '_station_description',
                          station_type + '_station_latitude',
                          station_type + '_station_longitude']].drop_duplicates(subset=[station_type + '_station_id'])

    # Merging with total count for each station
    popular_stations = pd.merge(popular_stations, unique_stations, on=station_type + '_station_id', how='inner')

    return popular_stations
```

```python
def popular_hours(date = None, df = hourly_rides):
    load_figure_template('BOOTSTRAP')
    
    if date is None:
        fig = px.line(
            df,
            x = 'started_at_hour',
            y = 'ride_count',
        )
        fig.update_layout(
            title = 'Most popular hours to rent a Bysykkel',
            yaxis_title = "Count",
            xaxis_title = "Hour on the day",
            xaxis=dict(showgrid=False),  # Remove x-axis gridlines
            yaxis=dict(showgrid=False)  # Remove y-axis gridlines
        )
        return fig
    # Convert to datetime object
    try:
        input_date = datetime.strptime(date, "%Y-%m")
    except ValueError:
        input_date = datetime.strptime(date, "%Y-%b")
        
    # Format the datetime object to the desired month format
    formatted_date = input_date.strftime("%Y-%m")
    
    # Filter the DataFrame for the specified month
    df_month = df[df['started_at'].dt.strftime('%Y-%m') == formatted_date]
    
    fig = px.line(
        df_month,
        x='started_at_hour',
        y='ride_count'
    )
    
    fig.update_layout(
        title=f'Hourly bysykkel rides for {formatted_date}',
        yaxis_title="Count",
        xaxis_title="Hour",
        xaxis=dict(showgrid=False),  # Remove x-axis gridlines
        yaxis=dict(showgrid=False)  # Remove y-axis gridlines
    )
    
    return fig
        
popular_hours()
```

```python
def get_best_day(month):
    # Convert month to datetime format
    month_date = pd.to_datetime(month)
    
    # Filter the daily_rides DataFrame for the specified month
    filtered_daily_rides = daily_rides[daily_rides['started_at'].dt.to_period('M') == month_date.to_period('M')]
    
    # Find the maximum ride count for the specified month
    max_ride_count = filtered_daily_rides['ride_count'].max()
    
    # Filter the data for the day(s) with the maximum ride count
    best_day = filtered_daily_rides[filtered_daily_rides['ride_count'] == max_ride_count]
    
    return best_day

get_best_day("2022-08")
```

```python
monthly_rides = pd.DataFrame(data.groupby('started_at_month').size()).reset_index()
monthly_rides = monthly_rides.rename(columns={0: 'total_ride_count'})
#avg_duration = pd.DataFrame(data.groupby('started_at_month')['duration'].mean()).reset_index()
monthly_rides

# Iterate over the rows of monthly_rides dataframe
for index, row in monthly_rides.iterrows():
    # Extract the month value from the row
    month = row['started_at_month']
    
    # Call the get_best_day function to get the best day for the month
    best_day = get_best_day(month)
    
    if not best_day.empty:
        # Get the maximum ride count and best day date
        max_ride_count = best_day['ride_count'].iloc[0]
        best_day_date = best_day['started_at'].dt.date.iloc[0]
    else:
        # If no best day found, set the values to NaN or any desired default value
        max_ride_count = None
        best_day_date = None
    
    # Assign the additional columns to the corresponding row
    monthly_rides.at[index, 'best_day_date'] = best_day_date
    monthly_rides.at[index, 'best_ride_count'] = max_ride_count

avg_duration = pd.DataFrame(data.groupby('started_at_month')['duration'].mean()).reset_index()
avg_duration = avg_duration.rename(columns={'duration': 'avg_duration'})

# Merge avg_duration into monthly_rides based on the 'started_at_month' column
monthly_rides = monthly_rides.merge(avg_duration, on='started_at_month', how='left')
monthly_rides
```

```python
def kpi_function(month = None):
    if month:
        # Convert month to standardized format
        month = pd.to_datetime(month).strftime('%Y-%m')
         
        filtered_month = monthly_rides[monthly_rides['started_at_month'] == month]
        
        if filtered_month.empty:
            return None
        
        # metrics for the specified month
        avg_duration_month = filtered_month['avg_duration'].item()
        total_rides_month = filtered_month['total_ride_count'].item()
        max_ride_count = filtered_month['best_ride_count'].item()
        best_day_date = filtered_month['best_day_date'].item()
        
        # Indicator
        fig_ind = go.Figure()
        fig_ind.add_trace(go.Indicator(
            mode="number",
            value=avg_duration_month,
            domain={'x': [0, 0.3], 'y': [0.99, 1]},
            title={"text": f"Average duration for {month}", "font": {"color": "blue"}},
            number_font={"size": 50},
            number={'suffix': ' sec'}
        ))
        fig_ind.add_trace(go.Indicator(
            mode="number",
            value=total_rides_month,
            delta={'reference': 200, 'relative': True},
            title={"text": f"Total amount of rides for {month}", "font": {"color": "blue"}},
            number_font={"size": 50},
            number={'valueformat': ',.0f'},
            domain={'x': [0.35, 0.65], 'y': [0.99, 1]}
        ))
        fig_ind.add_trace(go.Indicator(
            mode="number",
            value=max_ride_count,
            delta={'reference': 200, 'relative': True},
            title={"text": f"Best day of month - Amount of rides: {best_day_date}", "font": {"color": "blue"}},
            number_font={"size": 50},
            number={'valueformat': ',.0f'},
            domain={'x': [0.7, 1], 'y': [0.99, 1]}
        ))
        fig_ind.update_layout(height=100)
    else:
        # Average duration, total rides, maximum ride count and best month
        avg_duration_month = data['duration'].mean()
        
        total_rides_month = data['started_at'].count()
        
        max_ride_count = monthly_rides['total_ride_count'].max()
        
        best_month_count = monthly_rides['total_ride_count'].idxmax()
        
        best_month = monthly_rides.loc[best_month_count,'started_at_month']
        
        # Making the indicators
        fig_ind = go.Figure()
        fig_ind.add_trace(go.Indicator(
            mode="number",
            value=avg_duration_month,
            domain={'x': [0, 0.3], 'y': [0.99, 1]},
            delta={'reference': 100, 'relative': True},
            title={"text": "Average duration since May 2022", "font": {"color": "blue"}},
            number_font={"size": 50},
            number={'suffix': ' sec'}
        ))

        fig_ind.add_trace(go.Indicator(
            mode="number",
            value=total_rides_month,
            delta={'reference': 100, 'relative': True},
            title={"text": "Total amount of rides since May 2022", "font": {"color": "blue"}},
            number_font={"size": 50},
            number={'valueformat': ',.0f'},
            domain={'x': [0.35, 0.65], 'y': [0.99, 1]}
        ))

        fig_ind.add_trace(go.Indicator(
            mode="number",
            value=max_ride_count,
            delta={'reference': 100, 'relative': True},
            title={"text": f"Best month: {best_month}", "font": {"color": "blue"}},
            number_font={"size": 50},
            number={'valueformat': ',.0f'},
            domain={'x': [0.7, 1], 'y': [0.99, 1]}
        ))
        fig_ind.update_layout(height=100)

    return fig_ind

kpi_function("2022-08")
```

## Creating the app

```python
# Create the JupyterDash app
theme = dbc.themes.COSMO
app = JupyterDash(external_stylesheets=[theme])

# App layout
app.layout = dbc.Container(children=[
    html.H1('Oslo Bysykkel Dashboard', style={'textAlign': 'center'}),
    dcc.Markdown(description, style={'textAlign': 'center'}),
    dbc.Row(dbc.Col(
        dcc.Dropdown(
            id='month-dropdown',
            options=[
                {'label': f"{month.strftime('%B')}-{month.strftime('%Y')}", 'value': month.strftime('%Y-%m')}
                for month in pd.date_range(start='2022-05-01', end='2024-07-01', freq='MS')
            ],
            value=None,
            placeholder='Select a month'
        )), style={'width': '14%'}),
    dbc.Row(
        dbc.Col(
            dbc.Card(
                dcc.Graph(id='kpi',
                          className='mb-4',
                          style={'margin-bottom': '0'})
            ), style = {"height":"200px"}
        )
    ),
    dbc.Row([
        dbc.Col(dcc.Graph(id='daily_ride')),
        dbc.Col(dcc.Graph(id='daily_duration'))
    ], style={'margin-top': '0'}),
    dbc.Row([
        dbc.Col(dbc.Tabs([
            dbc.Tab(label='Start Station', tab_id='start-station'),
            dbc.Tab(label='End Station', tab_id='end-station'),
        ], id='station-tabs', active_tab='start-station')),
    ], style={'margin-top': '0'}),
    dbc.Row(children=[
        dbc.Col(html.Div(id='station-map-container',style ={'width':'100%'}))
    ], style = {'margin-bottom':'20px'})
])


@app.callback(
    Output('kpi', 'figure'),
    [Input('month-dropdown', 'value')]
)
def update_kpi_function(month):
    template = load_figure_template('COSMO')
    fig_kpi = kpi_function(month)
    return fig_kpi


@app.callback(
    Output('daily_ride', 'figure'),
    [Input('month-dropdown', 'value')]
)
def update_ride_count(date):
    template = load_figure_template('COSMO')
    fig_ride = ride_count(date)
    return fig_ride


@app.callback(
    Output('daily_duration', 'figure'),
    [Input('month-dropdown', 'value')]
)
def update_ride_duration(date):
    template = load_figure_template('COSMO')
    fig_duration = ride_duration(date)
    return fig_duration

@app.callback(
    Output('station-map-container', 'children'),
    [Input('station-tabs', 'active_tab'), Input('month-dropdown', 'value')]
)
def update_popular_stations(active_tab, month):
    # Retrieve a df based on the selected month
    if month is None:
        #df = popular_start_stations  # Assuming popular_start_stations is the default DataFrame
        if active_tab == 'start-station':
            df = popular_start_stations
        elif active_tab == 'end-station':
            df = popular_end_stations
        else:
            raise ValueError(f"Invalid active tab: {active_tab}")
    else:
    # Filter df on the selected month
        if active_tab == 'start-station':
            df = popular_start_stations_monthly[popular_start_stations_monthly['started_at_month'] == month]
        elif active_tab == 'end-station':
            df = popular_end_stations_monthly[popular_end_stations_monthly['started_at_month'] == month]
        else:
            raise ValueError(f"Invalid active tab: {active_tab}")

    map_fig = popular_stations(df)

    # Convert to HTML string
    map_html = map_fig._repr_html_()

    # Return html Iframe 
    return html.Iframe(srcDoc=map_html, style={'width': '100%', 'height': '500px'})

# run app
app.run_server(debug=True, port=8060)
```
